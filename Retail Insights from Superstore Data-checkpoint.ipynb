```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("1719219914-Analysis of Super Store - DA (1).csv")
df.head(5)
```

```python
df.drop(columns="Postal Code",inplace=True)
```

```python
df.describe()
```

```python
df.info()
```

```python
df.groupby("Region")["Sales"].sum().plot.pie(autopct="%1.0f%%")
```

```python
df.groupby("Segment")["Sales"].sum().plot.bar()
```

```python
# Plot sales and profit by category
fig, ax = plt.subplots(1, 2, figsize=(16,6))
df.groupby("Category")["Sales"].sum().plot.bar(ax=ax[0], color=['blue', 'orange', 'green'])
ax[0].set_title('Total Sales by Category')
ax[0].set_xlabel('Category')
ax[0].set_ylabel('Total Sales')

df.groupby("Category")["Profit"].sum().plot.bar(ax=ax[1], color=['blue', 'orange', 'green'])
ax[1].set_title('Total Profit by Category')
ax[1].set_xlabel('Category')
ax[1].set_ylabel('Total Profit')

plt.show()
```

```python
# Plot profit by region
plt.figure(figsize=(10,6))
df.groupby("Region")["Profit"].sum().plot.bar(color=['blue', 'orange', 'green', 'red'])
plt.title('Total Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')
plt.show()
```

```python
# Plot profit by segment
plt.figure(figsize=(10,6))
df.groupby("Segment")["Profit"].sum().plot.bar(color=['blue', 'orange', 'green'])
plt.title('Total Profit by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Profit')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df, hue='Segment', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Features', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title('Box Plot of Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x='Category', y='Profit', data=df)
plt.title('Violin Plot of Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Ship Mode', data=df)
plt.title('Count Plot of Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(14,8))
df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).plot.bar()
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
plt.figure(figsize=(14,8))
df.groupby("State")["Profit"].sum().sort_values(ascending=False).plot.bar(color='purple')
plt.title('Total Profit by State')
plt.xlabel('State')
plt.ylabel('Total Profit')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Sales', y='Profit', hue='Category', data=df)
plt.title('Scatter Plot of Sales vs. Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Ship Mode', data=df)
plt.title('Count Plot of Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Count')
plt.show()
```
